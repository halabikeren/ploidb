```python
import requests
from requests.models import Response
from collections import defaultdict
from bs4 import BeautifulSoup
import requests
from time import sleep
import pandas as pd
import os
```

```python
FAIL_SEEP_TIME = 120
SUCCESS_SLEEP_TIME = 5
MAX_REQUESTS = 300
RETRIES_NUM_LIMIT = 10

req_address = "https://cvalues.science.kew.org/search"
req_headers = {
    "accept": "text/html,application/xhtml+xml,application/xml;q=0.9,image/avif,image/webp,image/apng,*/*;q=0.8,application/signed-exchange;v=b3;q=0.9",
    "accept-language": "en-US,en;q=0.9,he;q=0.8",
    "cache-control": "max-age=0",
    "content-type": "application/x-www-form-urlencoded",
    "sec-ch-ua": "\"Chromium\";v=\"104\", \" Not A;Brand\";v=\"99\", \"Google Chrome\";v=\"104\"",
    "sec-ch-ua-mobile": "?0",
    "sec-ch-ua-platform": "\"Windows\"",
    "sec-fetch-dest": "document",
    "sec-fetch-mode": "navigate",
    "sec-fetch-site": "same-origin",
    "sec-fetch-user": "?1",
    "upgrade-insecure-requests": "1"
  }
req_body = "search_type=AllPlants&prime_estimates=1&show_fields%5B%5D=family&show_fields%5B%5D=genus&show_fields%5B%5D=species&show_fields%5B%5D=subspecies&show_fields%5B%5D=authority&show_fields%5B%5D=chromosome&show_fields%5B%5D=ploidy&show_fields%5B%5D=estimation&show_fields%5B%5D=voucher&show_fields%5B%5D=prime&show_fields%5B%5D=reference&cvalue=onec&family=&genus=&species=&dna_from=&dna_to=&chromosome_from=&chromosome_to=&ploidy_from=&ploidy_to=&voucher=&estimation_method=&angiosperm_group=&angiosperm_lifecycle=&gymnosperm_group=&sperm_flagella_number=&algae_group=&algae_order=&bryophyte_group=&pteridophyte_group=&pteridophyte_spore_type=&pteridophyte_sporangium_type=&sort=cf.fourc&sort_order=ASC{page}"

db_output_path=f"../chromevol/results/cvalues_db.csv"
```

```python
def parse_response(resp: requests.models.Response) -> dict:
    soup=BeautifulSoup(resp.text)
    data = dict()
    headers = []
    for index,header in  enumerate(header.get_text().strip() for header in soup.find("table",{"class":"search-results"}).find("thead").find_all("th")):
        data[header] = []
        headers.append(header)
    for row in soup.find("table",{"class":"search-results"}).find("tbody").find_all("tr"):
        for index,cell in enumerate(row.find_all("td")):
            data[headers[index]].append(cell.get_text())
    return data
    

dfs = []
requests_count = 0
failed_pages = []
for page_index in range(1, 124):
    resp=requests.post(req_address,headers=req_headers,data=req_body.format(page=f"&page={page_index}" if page_index > 1 else ""))
    print(f"page_index={page_index} recieved status code {resp.status_code}")
    requests_count += 1
    if requests_count > MAX_REQUESTS:
        break
    retries = 0
    while resp.status_code != 200:
        if retries > RETRIES_NUM_LIMIT:
            failed_pages.append(page_index)
            continue
        sleep(FAIL_SEEP_TIME)
        resp=requests.post(req_address,headers=req_headers,data=req_body.format(page=f"&page={page_index}" if page_index > 1 else ""))
        retries += 1
        requests_count += 1
        if requests_count > MAX_REQUESTS:
            break
    data = parse_response(resp)
    dfs.append(pd.DataFrame(data))
    sleep(SUCCESS_SLEEP_TIME)
failed_pages += list(range(page_index,124))
failed_pages = list(set(failed_pages))
print(f"#failed_pages={len(failed_pages)}")
db = pd.concat(dfs)
db.to_csv(db_output_path, index=False)
```
